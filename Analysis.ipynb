# Comparative Analysis

```python
# Applying moving average to stock_price and index_point column to smoothen the graph
window_size = 30
df['stock_price_smooth'] = df['stock_price'].rolling(window=window_size).mean()
df['index_point_smooth'] = df['index_point'].rolling(window=window_size).mean()

fig = make_subplots(specs=[[{"secondary_y": True}]])

# left y-axis (for stock price)
fig.add_trace(go.Scatter(x=df['date'], y=df['stock_price_smooth'], mode='lines', line_shape='linear',
                         name='Reliance Stock Price', line_color='blue'), secondary_y=False)

# Right y-axis (for index point)
fig.add_trace(go.Scatter(x=df['date'], y=df['index_point_smooth'], mode='lines', line_shape='linear',
                         name='Nifty 50 Index Point', line_color='orange'), secondary_y=True)

fig.update_yaxes(title_text='Reliance Stock Price (â‚¹)', secondary_y=False)
fig.update_yaxes(title_text='Nifty 50 Index Point', secondary_y=True)
fig.update_xaxes(title_text='Date')
fig.update_layout(title_text='Reliance Stock Price vs. Nifty 50 Index Point Movement Over Time')
fig.update_layout(plot_bgcolor='#e6ffe6')
fig.update_xaxes(showgrid=True, gridcolor='white', gridwidth=1)
fig.update_yaxes(showgrid=True, gridcolor='white', gridwidth=1)

fig.show()
```

1) Most of the time stock price and index point follows same trend with few exceptions.
2) Before 2020, we can see one inteval (Mar 2015) where the stock price underperformed the index point.
3) After 2020, we can see two intevals (Oct 2020, May 2022) where stock price outperformed the index point.

```python
# Applying moving average to stock_volume and index_volume column to smoothen the graph
window_size = 50
df['stock_price_smooth'] = df['stock_price'].rolling(window=window_size).mean()
df['stock_volume_smooth'] = df['stock_vol'].rolling(window=window_size).mean()

fig = make_subplots(specs=[[{"secondary_y": True}]])

# left y-axis (for stock volume)
fig.add_trace(go.Scatter(x=df['date'], y=df['stock_price_smooth'], mode='lines', line_shape='linear',
                         name='Reliance Stock Price', line_color = 'blue'), secondary_y=False)

# Right y-axis (for index volume)
fig.add_trace(go.Scatter(x=df['date'], y=df['stock_volume_smooth'], mode='lines', line_shape='linear',
                         name='Reliance Stock Volume', line_color = 'orange'), secondary_y=True)

fig.update_yaxes(title_text='Reliance Stock Price', secondary_y=False)

fig.update_yaxes(title_text='Reliance Stock Volume', secondary_y=True)

fig.update_xaxes(title_text='Date')

fig.update_layout(title_text='Reliance Stock Price vs. Volume Movement Over Time')

fig.update_layout(plot_bgcolor='#e6ffe6')
fig.update_xaxes(showgrid=True, gridcolor='white', gridwidth=1)
fig.update_yaxes(showgrid=True, gridcolor='white', gridwidth=1)

fig.show()
```

After Sept 2020, The stock price is steadily increasing but there's steep downfall in stock volume.

# Alpha Beta Analysis

```python
# Calculating daily returns
df['stock_return'] = df['stock_price'].pct_change()
df['index_return'] = df['index_point'].pct_change()

# Calculating covariance between stock returns and market returns
covariance = np.cov(df['Stock Returns'].dropna(), df['Market Returns'].dropna())[0][1]

# Calculate variance of market returns
variance_market = np.var(df['Market Returns'].dropna())

# Calculate beta
beta = round(covariance / variance_market, 2)

print("Volatility Indicator (Beta)", beta)
print("Reliance stock in question tends to be slightly more volatile than the overall market.")
```

```python
# Calculating alpha
risk_free_rate = 7.16 
market_return = df['index_return'].mean() 
beta = 1.09  

df['Alpha'] = df['stock_return'] - (risk_free_rate + beta * (market_return - risk_free_rate))

alpha = round(df['Alpha'].mean(), 2)

print("Return Indicator (Alpha):", alpha)
print("Reliance stock has performed better than what would be expected based on the overall market trends.")
```
